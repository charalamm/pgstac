```python
# %pip install pystac_client orjson smart_open
from pystac_client import Client
import orjson
from smart_open import open

catalog = Client.open('https://planetarycomputer.microsoft.com/api/stac/v1')
for collection in catalog.get_collections():
    if collection.id == 'modis-13Q1-061':
        with open('collections.ndjson.gz','wb') as collection_ndjson:
            collection_ndjson.write(orjson.dumps(collection.to_dict()))
        items=collection.get_items()
        with open('items.ndjson.gz', 'wb') as items_ndjson:
            for i in range(10000):
                item = next(items)
                print(item.id)
                items_ndjson.write(orjson.dumps(item.to_dict()))
                items_ndjson.write(b'\n')
```
